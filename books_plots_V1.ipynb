```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
books = "output/combined_book.csv"
books_df = pd.read_csv(books)
books_df.head()
```

```python
data_new1 = books_df.copy()                              
data_new1 = data_new1.replace(r'^s*$', float('NaN'), regex = True)  
data_new1_df = pd.DataFrame(data_new1)
data_new1_df
```

```python
data_new2 = data_new1.copy()                          
data_new2.dropna(inplace = True)                     
data_new2_df = pd.DataFrame(data_new2)
data_new2_df
```

```python
data_new2.columns
```

```python
#Top Rating Books
top_authors = data_new2.groupby(["category", "title", "author"])
top_authors.count()

top_group_count = top_authors["Amazon Rating Total"].count()


top_df = pd.DataFrame(top_group_count)

top_df.sort_values(by="title", ascending=False).head()
```

```python
#Top Rating Authors 
top_authors = data_new2.groupby(["category", "author"])
top_authors.count()

top_group_count = top_authors["Amazon Rating Total"].count()


top_df = pd.DataFrame(top_group_count)

top_df.sort_values(by='author', ascending=False).head()
```

```python
price_rating_col = data_new2.loc[:,['category','Amazon Rating Total','Amazon Price']]
price_rating_col
```

```python
book_group = price_rating_col.groupby("category")
#book_group_count = book_group["primary_isbn13"].nunique()
avg_price = book_group["Amazon Price"].mean()
#avg_rating = book_group["Amazon Rating Total"].mean()

price_vs_rating = pd.DataFrame({
    "Price Average": avg_price,
    #"Rating Average": avg_rating
})
#price_vs_rating.index.name= None
price_vs_rating
```

```python
# Price Average per Categories 
price_vs_rating.plot(kind="bar", figsize=(20,3))

# Set a title for the chart
plt.title("Price Average per Categories")

plt.show()
plt.tight_layout()
```
